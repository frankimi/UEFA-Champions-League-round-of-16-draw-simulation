```python
NO1_team_list = [('PSG','France','A'),
                 ('Bayern', 'Germany','B'),
                 ('Man City','England','C'),
                 ('Juventus','Italy','D'),
                 ('Liverpool','England','E'),
                 ('Barcelona','Spain','F'),
                 ('Leipzig','Germany','G'),
                 ('Valencia','Spain','H')
                ]

NO2_team_list = [('Real Madrid','Spain','A'),
                 ('Tottenham', 'England','B'),
                 ('Atalanta','Italy','C'),
                 ('Atletico Madrid','Spain','D'),
                 ('Napoli','Italy','E'),
                 ('Dortmund','Germany','F'),
                 ('Lyon','France','G'),
                 ('Chelsea','England','H')
                ]
```

```python
import pandas as pd
from random import randrange
```

```python
# Create dataframe to store results and initialize

sum_results = pd.DataFrame(columns = [x[0] for x in NO1_team_list], index = [x[0] for x in NO2_team_list])
for team in [x[0] for x in NO1_team_list]:
    sum_results[team] = 0

sum_results
```

```python
# One single draw


NO1_team_list_draw  = NO1_team_list.copy()
NO2_team_list_draw  = NO2_team_list.copy()


result = []
while len(NO1_team_list_draw) != 0 and len(NO2_team_list_draw) != 0:
    selected_NO1_team = NO1_team_list_draw.pop(randrange(len(NO1_team_list_draw)))
    country = selected_NO1_team[1]
    group = selected_NO1_team[2]

    NO2_team_to_be_selected = [team for team in NO2_team_list_draw if team[1] != country and team[2] != group]
    selected_NO2_team = NO2_team_to_be_selected[randrange(len(NO2_team_to_be_selected)-1)]

    result = result + [(selected_NO1_team,selected_NO2_team)]


print(result)
```

```python


sum_results = pd.DataFrame(columns = [x[0] for x in NO1_team_list], index = [x[0] for x in NO2_team_list])
for team in [x[0] for x in NO1_team_list]:
    sum_results[team] = 0

# set max draw of 100 times

total_valid_draws = 0

while total_valid_draws < 100:

    NO1_team_list_draw  = NO1_team_list.copy()
    NO2_team_list_draw  = NO2_team_list.copy()

    result = []
    while len(NO1_team_list_draw) != 0 and len(NO2_team_list_draw) != 0:
        selected_NO1_team = NO1_team_list_draw.pop(randrange(len(NO1_team_list_draw)))
        country = selected_NO1_team[1]
        group = selected_NO1_team[2]

        NO2_team_to_be_selected = [team for team in NO2_team_list_draw if team[1] != country and team[2] != group]
        if len(NO2_team_to_be_selected) >= 1:
            selected_NO2_team = NO2_team_to_be_selected[randrange(len(NO2_team_to_be_selected))]
            NO2_team_list_draw.remove(selected_NO2_team)

            result = result + [(selected_NO1_team[0], selected_NO2_team[0])]
           
    if len(result) == 8:
        for draw in result:
            sum_results[draw[0]][draw[1]] += 1
        total_valid_draws += 1

sum_results
```
